```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("BostonHousing.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['medv'])
y = df['medv']
cols = X.columns

scaler = StandardScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X,columns=cols)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

print('X_train:',X_train.shape)
print('y_train:',y_train.shape)
print('X_test:',X_test.shape)
print('y_test:',y_test.shape)
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

module = Sequential()
module.add(layers.Dense(32,activation = "sigmoid",input_shape = (X_train.shape[1],)))
module.add(layers.Dense(1,activation = 'linear'))
module.summary()
```

```python
module.compile(optimizer='sgd',loss='mean_squared_error')
module.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20)
```

```python
preds = module.predict(X_test).reshape(-1)

print(preds.shape)
print(preds)
```

```python
df1 = pd.DataFrame()
df1["actual"] = y_test.values
df1["preds"]  = preds
df1
```
